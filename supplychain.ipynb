```python
# for data manipulation and analysis
import pandas as pd

# Writing to a CSV File
import csv

# for better display
from IPython.display import display

# for vizualization
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import seaborn as sns

# Numerical computing
import numpy as np

import math

from deep_translator import GoogleTranslator

import us 

```

```python
def wrangle(filepath):
    # Load data
    df = pd.read_excel(filepath)

    return df
```

```python
filepath = r"C:\Users\KINGLEY EYO\Documents\supply_chain_sql_clean.xlsx"
df = wrangle(filepath)
```

```python
df.head()
```

```python
cols_to_translate = ['Customer City', 'Customer Country', 'Order Country', 'Order State', 'Order City']

for col in cols_to_translate:
    print(f"Translating column: {col}")
    
    unique_vals = df[col].dropna().unique()
    translation_map = {}
    
    for val in unique_vals:
        try:
            translated = GoogleTranslator(source='auto', target='en').translate(val)
            translation_map[val] = translated
        except Exception as e:
            print(f"Error translating '{val}': {e}")
            translation_map[val] = val  # fallback: keep original
    
    df[col] = df[col].map(translation_map)
```
